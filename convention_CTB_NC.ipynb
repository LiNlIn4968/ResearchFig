```python
import pandas as pd
import matplotlib.pyplot as plt

# --- 1. IEEE 标准学术风格设置 ---
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['DejaVu Serif']
plt.rcParams['font.size'] = 9                  # 适合双栏排版的字号
plt.rcParams['axes.labelsize'] = 9
plt.rcParams['legend.fontsize'] = 8
plt.rcParams['xtick.labelsize'] = 8
plt.rcParams['ytick.labelsize'] = 8



# 1. 读取数据
# 假设文件名为 'noCTB_CTB.xlsx'
file_path = 'noCTB_CTB.xlsx'
df = pd.read_excel(file_path)

# 2. 提取数据
# 第一组数据 (1-4列，索引0-3)
freq1 = df.iloc[:, 0]
gain1 = df.iloc[:, 1]
phase1 = df.iloc[:, 3]

# 第二组数据 (6-9列，索引5-8)
freq2 = df.iloc[:, 5]
gain2 = df.iloc[:, 6]
phase2 = df.iloc[:, 8]

# 3. 开始绘图
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(5, 4), sharex=True)

# --- 上图：增益曲线 (Gain) ---
ax1.semilogx(freq1, gain1, label='Conventional Pseudo-Resistor', color='#4054a1', linewidth=2.2)
ax1.semilogx(freq2, gain2, label='CTB Pseudo-Resistor', color='#d93e32', linestyle='--', linewidth=2.2)
ax1.set_ylabel('Gain (dB)')
ax1.set_title('Bode Plot of Proposed Neural Amplifier')
ax1.grid(True, which="both", ls="-", alpha=0.5)
ax1.legend()

# --- 下图：相位曲线 (Phase) ---
ax2.semilogx(freq1, phase1, color='#4054a1', linewidth=2.2)
ax2.semilogx(freq2, phase2, color='#d93e32', linestyle='--', linewidth=2.2)
ax2.set_ylabel('Phase (Degree)')
ax2.set_xlabel('Frequency (Hz)')
ax2.grid(True, which="both", ls="-", alpha=0.5)

# 设置范围
ax1.set_xlim([1e-4, 1e6])
# ax2.set_ylim(top=-80)

# 调整布局
plt.tight_layout()
plt.show()
```
